Now, we load all days for January 2007

```python
def LoadMonthlyData(path):
    Data = []
    days_in_month = len(os.listdir(path)) -1
    for day in range(1, days_in_month):
        if day < 10:
            fullpathnameJan = path + '0' + str(day) + '/'
        else:
            fullpathnameJan = path + str(day) + '/'
        MeasurementMoments = os.listdir(fullpathnameJan)
        for file in MeasurementMoments:
            filename = fullpathnameJan + file
            NPData = np.loadtxt(filename, dtype='str')
            PDData = pd.DataFrame(NPData, columns = columns)
            Data.append(PDData)
    return Data
```
